```python
jobs = []
companies = []
locations = []
keywords = ['data%20engineer', 'Devops', 'Software%20Developer']
base_url = 'https://www.linkedin.com/jobs/search/?currentJobId=3528775050&distance=25&f_TPR=r86400&geoId=102890719&keywords={}&location=Netherlands&origin=JOB_SEARCH_PAGE_JOB_FILTER&start={}'
links = []
timepoints = []

for word in keywords:
    page_still_valid = True
    page=0  
    while page_still_valid:
        client = ScrapflyClient(key="scp-live-97e66d928428429c8ae1f3d0d9d2a352")
        result = client.scrape(ScrapeConfig(
        url=base_url.format(word,page),
        asp=True,
                # ^ enable Anti Scraping Protection
        ))
          

        soup = bs4.BeautifulSoup(result.content, "lxml")
        job_titles = soup.find_all(class_="base-search-card__title")
        company_titles = soup.find_all(class_="hidden-nested-link")
        location = soup.find_all(class_="job-search-card__location")
        job_link = soup.find_all(class_="base-card__full-link absolute top-0 right-0 bottom-0 left-0 p-0 z-[2]")
        job_times = soup.find_all("time")

        for title in job_titles:
            jobs.append(title.text)
        jobs = [job.replace("\n", "") for job in jobs]
        
        for title1 in company_titles:
            companies.append(title1.text)
        companies = [company.replace("\n", "") for company in companies]

        for title2 in location:
            locations.append(title2.text)
        locations = [location.replace("\n", "") for location in locations]
        
        for link in job_link:
            links.append(link.get('href'))

        for job_time in job_times:
            time_text = job_time.get_text(strip=True)
            timepoints.append(time_text)

        
        if "You've viewed all jobs for this search" in result.content:
            break
        page = page+25
     
```

```python
result = pd.DataFrame()
result = pd.DataFrame({'Companies': companies,'Jobs':jobs, 'Locations': locations, 'Posted':timepoints, 'Link' : links})
result = result.drop_duplicates()

result
```
